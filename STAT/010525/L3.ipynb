Kicking horses

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

#data
deaths = np.array([0,1,2,3,4])
groups = np.array([109,65,22,3,1])

tot = np.sum(groups)

f = groups/tot

plt.scatter(deaths, f)
plt.plot(deaths, scipy.stats.poisson.pmf(deaths, 0.5), label='$\mu = 0.5$')
plt.plot(deaths, scipy.stats.poisson.pmf(deaths, 0.1), label='$\mu = 0.1$')
plt.plot(deaths, scipy.stats.poisson.pmf(deaths, 1.1), label='$\mu = 1.1$')
plt.plot(deaths, scipy.stats.poisson.pmf(deaths, np.average(deaths, weights=f)), label='$\mu = x_{ave} =$' + str(f'{np.average(deaths, weights=f):.2f}') + '')

plt.legend()
plt.xlabel('Deaths')
plt.ylabel('freq')
```
